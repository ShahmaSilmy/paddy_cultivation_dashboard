```python
import pandas as pd

# Load each file and add Year + Season columns manually
df1 = pd.read_csv('paddy_2020_maha_raw.csv')
df1['Year'] = 2020
df1['Season'] = 'Maha'

df2 = pd.read_csv('paddy_2020_yala_raw.csv')
df2['Year'] = 2020
df2['Season'] = 'Yala'

df3 = pd.read_csv('paddy_2021_maha_raw.csv')
df3['Year'] = 2021
df3['Season'] = 'Maha'

df4 = pd.read_csv('paddy_2021_yala_raw.csv')
df4['Year'] = 2021
df4['Season'] = 'Yala'

df5 = pd.read_csv('paddy_2022_maha_raw.csv')
df5['Year'] = 2022
df5['Season'] = 'Maha'

df6 = pd.read_csv('paddy_2022_yala_raw.csv')
df6['Year'] = 2022
df6['Season'] = 'Yala'

df7 = pd.read_csv('paddy_2023_maha_raw.csv')
df7['Year'] = 2023
df7['Season'] = 'Maha'

df8 = pd.read_csv('paddy_2023_yala_raw.csv')
df8['Year'] = 2023
df8['Season'] = 'Yala'
```

```python
# Combine all 8 dataframes
df_all = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8], ignore_index=True)
```

```python
# Save combined dataset
df_all.to_csv('paddy_2020_2023_combined.csv', index=False)

# Download it
files.download('paddy_2020_2023_combined.csv')
```
